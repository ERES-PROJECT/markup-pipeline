```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("markup_2017_01_to_2024_08.csv")

df["Date"] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Month"].astype(str))

plt.figure(figsize=(10,6))

for edc in df["EDC"].unique():
    subset = df[df["EDC"] == edc]
    plt.plot(subset["Date"], subset["markup_da"], label=edc)

plt.legend()
plt.title("Markup (DA) Over Time by EDC")
plt.xlabel("Time")
plt.ylabel("Markup")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
df.groupby("EDC")["markup_da"].mean().plot(kind="bar")

plt.title("Average Markup by EDC")
plt.ylabel("Markup")
plt.tight_layout()
plt.show()
```
